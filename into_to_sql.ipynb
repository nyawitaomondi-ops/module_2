```python
import sqlite3
```

```python
#connection to database
conn = sqlite3.connect(':memory:')
#create the cursor
cursor = conn.cursor()
```

```python
c1 =  """
CREATE TABLE students(
    student_ID INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER, 
    city TEXT,
    email TEXT,
    gender TEXT 
)
"""

cursor.execute(c1)
```

```python
query_1 = """
    SELECT 
        * 
    FROM students;"""

cursor.execute(query_1)
cursor.fetchall()
```

```python
query_2 = """
    SELECT 
        name, city 
    FROM students;"""

cursor.execute(query_2)
cursor.fetchall()
```

```python
#Filtering
query_3 = """
    SELECT 
        name, city 
    FROM students
    WHERE 
        city = 'Seattle';"""

cursor.execute(query_3)
cursor.fetchall()
```

```python
#Filtering
query_4 = """
    SELECT 
        name, city 
    FROM students
    WHERE 
        age BETWEEN 20 AND 25;"""

cursor.execute(query_4)
cursor.fetchall()
```

```python
# Ordering (ASC or DESC)
query_5 = """
    SELECT 
        name, city, age 
    FROM students
    ORDER BY age DESC;"""

cursor.execute(query_5)
cursor.fetchall()
```
